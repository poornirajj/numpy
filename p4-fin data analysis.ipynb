```python
import numpy as np


np.random.seed(0)  
stock_prices = np.random.randint(100, 501, size=(5, 12))  # 5 stocks Ã— 12 months

print("Stock Prices (USD):\n", stock_prices)


percentage_change = np.diff(stock_prices, axis=1) / stock_prices[:, :-1] * 100
print("\nPercentage Change (%):\n", np.round(percentage_change, 2))

row_mean = stock_prices.mean(axis=1, keepdims=True)
normalized_prices = stock_prices - row_mean
print("\nNormalized Prices (Mean Subtracted):\n", np.round(normalized_prices, 2))


volatility = np.sqrt(np.var(stock_prices, axis=1))
print("\nVolatility (Std Dev via sqrt(var)):\n", np.round(volatility, 2))


exp_growth = np.exp(stock_prices / 100)  # scaled to avoid overflow
print("\nExponential Price Growth:\n", np.round(exp_growth, 2))

log_prices = np.log10(stock_prices)
print("\nLog10 of Stock Prices:\n", np.round(log_prices, 2))


mean_prices = np.mean(stock_prices, axis=1)
median_prices = np.median(stock_prices, axis=1)
max_prices = np.max(stock_prices, axis=1)
min_prices = np.min(stock_prices, axis=1)
variance_prices = np.var(stock_prices, axis=1)
std_dev_prices = np.std(stock_prices, axis=1)

print("\nMean Prices per Stock:", np.round(mean_prices, 2))
print("Median Prices per Stock:", np.round(median_prices, 2))
print("Max Prices per Stock:", max_prices)
print("Min Prices per Stock:", min_prices)
print("Variance per Stock:", np.round(variance_prices, 2))
print("Standard Deviation per Stock:", np.round(std_dev_prices, 2))
```
